```python
from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale = 1./255,shear_range = 0.2,zoom_range = 0.2,horizontal_flip = True)
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
from google.colab import drive      #COMMAND FOR GOOGLE MOUNT
drive.mount('/content/drive')
```

```python
x_train = train_datagen.flow_from_directory('/content/drive/MyDrive/dataset 1/train',target_size=(75,75),batch_size=32,class_mode='binary')
x_test = test_datagen.flow_from_directory('/content/drive/MyDrive/dataset 1/test',target_size=(75,75),batch_size=32,class_mode='binary')
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Flatten
```

```python
model = Sequential()
```

```python
model.add(Conv2D(64,(3,3), activation='relu', input_shape=(75,75,3)))
```

```python
model.add(MaxPooling2D(pool_size=(3,3), strides=(2,2)))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(units= 512, kernel_initializer="uniform", activation="relu"))
model.add(Dense(units = 256, kernel_initializer="uniform", activation="relu"))
model.add(Dense(units = 1, kernel_initializer="uniform", activation="sigmoid"))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit_generator(x_train,steps_per_epoch=22, epochs= 10, validation_data=x_test, validation_steps=20)
```
